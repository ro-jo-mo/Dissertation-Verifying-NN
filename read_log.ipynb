```python
import pickle
def read(path):
    with open(path,"rb") as f:
        abc = pickle.load(f)
        for _,var in abc.items():
            total = len(var["unsafe-pgd"]) + len(var["unknown"]) + len(var["safe"]) + len(var["safe-incomplete"])
            verified = len(var["safe"]) + len(var["safe-incomplete"])
            print(f"Verified Percentage: {100 * verified / total:.1f}")
            for a,b in var.items():
                print(f"{a} : {b}")
            break
        


```
